```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def func(x,y):
    return x**2 + y**2
```

```python
x = np.linspace(-0.7, 0.7, 150)
y = np.linspace(-0.7, 0.7, 150)
a, b = np.meshgrid(x,y)
C = a**2 + b**2 - 0.2

figure, axes = plt.subplots()

axes.contour(a,b,C, [0])
axes.set_aspect(1)
plt.show()
```
